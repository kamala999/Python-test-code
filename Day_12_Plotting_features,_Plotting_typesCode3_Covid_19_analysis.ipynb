```python

import matplotlib.pyplot as plt

sub = ["Telugu", "English", "Hindi", "Maths", "Social Science"]
s1 = [35, 45, 55, 77, 88]
s2 = [37, 47, 57, 74, 78]
cut = [70, 70, 70, 70, 70]

# Plotting
plt.scatter(sub, s1, label="Student 1", color="k", marker="P")
plt.scatter(sub, s2, label="Student 2", color="b", marker="H")
plt.plot(sub, cut, label="Cutoff", linestyle=":", linewidth=3)

plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.legend()
plt.show()
```

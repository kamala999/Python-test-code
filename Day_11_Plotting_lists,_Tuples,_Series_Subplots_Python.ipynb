```python
import matplotlib.pyplot as plt

overs = [1, 2, 3, 4, 5]
ind = [6, 12, 13, 18, 20]
aus = [5, 10, 15, 17, 18]

bar_width = 0.35  # Width of each bar

# Calculate the position of bars for India and Australia
bar_positions_ind = [x - bar_width/2 for x in overs]
bar_positions_aus = [x + bar_width/2 for x in overs]

# Plotting the bars for India and Australia
plt.bar(bar_positions_ind, ind, width=bar_width, label="India")
plt.bar(bar_positions_aus, aus, width=bar_width, label="Australia")

# Adding labels, legend, and grid
plt.xlabel('Overs')
plt.ylabel('Scores')
plt.title('Comparison of India and Australia Scores')
plt.xticks(overs)
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Subplots ---comparission bitween two team scores by using bar graphs
# importing library
import matplotlib.pyplot as plt # accessing -->plt.fun()
overs = [1,2,3,4,5]
ind = [6,12,13,18,20]
aus = [5,10,15,17,18]
plt.subplots(1,2)
plt.subplot(1,2,1)
plt.bar(overs,ind,label="india")#bar grapgs
plt.legend()
plt.grid()
plt.subplot(1,2,2)
plt.bar(overs,aus,label="Australia")
plt.legend()
plt.grid()# grid is using for boxes
```
